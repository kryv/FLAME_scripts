```python
from flame_utils import ModelFlame, hplot
import matplotlib.pyplot as plt
```

## create `ModelFlame` object from lattice file

```python
fm = ModelFlame('PS_demo.lat')
```

## plot xrms and yrms

```python
hplot('xrms', 'yrms', machine=fm)
```

## clone machine and create new `ModelFlame` as `fm2`

```python
fm2 = ModelFlame(machine=fm.clone_machine())
```

## get list of sbend elements

```python
sb = fm2.find(type='sbend')
print(sb)
```

## remove all sbend from `fm2`

```python
fm2.pop_element(fm2.find(type='sbend'))
```

## plot xrms and yrms of `fm2` (green boxes (=sbend) are removed)

```python
hplot('xrms', 'yrms', machine=fm2)
```

## no changes in `fm`

```python
hplot('xrms', 'yrms', machine=fm)
```

## e.g. remove 'aper' attribute from element

```python
d = fm.get_element(index=4)[0]['properties']
print(d)
d.pop('aper')
fm.pop_element(index=4)
fm.insert_element(index=4, element=d)

d2 = fm.get_element(index=4)[0]['properties']
print(d2)
```
